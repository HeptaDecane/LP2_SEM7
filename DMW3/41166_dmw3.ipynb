```python
import numpy as np 
import pandas as pd 
from apyori import apriori 
```

```python
basket_data = pd.read_csv('market-basket.csv')
basket = pd.DataFrame(basket_data)
```

```python
basket
```

```python
basketContents = []
for i in range(0, 1000):
    basketContents.append([str(basket.values[i,j]) for j in range(1, 33)])
```

```python
association_rules = apriori(basketContents, min_support=0.015, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
print(association_results)
```

```python
df_results = pd.DataFrame(association_results)
```

```python
first_values = []
second_values = []
third_values = []
fourth_value = []

for i in range(df_results.shape[0]):
    single_list = df_results['ordered_statistics'][i][0]
    first_values.append(list(single_list[0]))
    second_values.append(list(single_list[1]))
    third_values.append(single_list[2])
    fourth_value.append(single_list[3])
    
# support = list(df_results['support'])
support = pd.DataFrame(df_results['support'])
```

```python
lhs = pd.DataFrame(first_values)
rhs= pd.DataFrame(second_values)
confidence=pd.DataFrame(third_values,columns=['confidence'])
lift=pd.DataFrame(fourth_value,columns=['lift'])
```

```python
df_final = pd.concat([lhs,rhs,support,confidence,lift], axis=1)
```

```python
df_final.fillna(value=' ', inplace=True)
df_final.columns = ['lhs',0,1,2,'rhs','support','confidence','lift']
```

```python
df_final['lhs'] = df_final['lhs']+str(", ")+df_final[1]+str(", ")+df_final[2]
```

```python
df_final.columns = ['items_base','items_add','support','confidence','lift']
```
